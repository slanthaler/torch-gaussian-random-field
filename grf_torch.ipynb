This notebook illustrates sampling of GRFs, using the implemented classes in grf.py.
This implementation employs an exponential decay of the Fourier coefficients.

### GRF in one dimension

```python
import torch
import matplotlib.pyplot as plt

from grf import GRF1d

#
length_scale = 0.5
alpha = 1.0
tau = 2.0
grf = GRF1d(length_scale=length_scale, alpha=alpha, tau=tau)
```

```python
# produce a few samples on an equidistant grid
Nsample = 4
Ngrid = 200
u = grf.sample(Nsample, Ngrid)
grid = grf.grid(Ngrid)
u.shape
```

```python
plt.figure(figsize=(15,3))
for i, sample in enumerate(u):
    plt.subplot(1,u.shape[0],i+1)
    plt.plot(grid, sample)
    plt.grid()
    plt.title(f'Sample {i}')
plt.show()
```

### GRF in two dimensions

```python
from grf import GRF2d

#
length_scale = 0.5
alpha = 1.0
tau = 2.0
grf = GRF2d(length_scale=length_scale, alpha=alpha, tau=tau)
```

```python
# produce a few samples on an equidistant grid
Nsample = 4
Ngrid = 100
u = grf.sample(Nsample, Ngrid)
X,Y = grf.grid(Ngrid)
u.shape
```

```python
plt.figure(figsize=(15,3))
for i, sample in enumerate(u):
    plt.subplot(1,u.shape[0],i+1)
    plt.pcolor(X, Y, sample)
    plt.colorbar()
    plt.title(f'Sample {i}')
    plt.tight_layout()
plt.show()
```
